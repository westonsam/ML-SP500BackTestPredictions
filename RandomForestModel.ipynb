# Initial Baseline Model - Random Forest Classifier

Why is this a good initial model?
Random Forests use a lot of individual decision trees with randomized parameters and then average the results from those decision trees. 
As a result the model:
is resistant to overfitting
runs relatively quickly
can pick up non-linear tendencies in the data

```python
from sklearn.ensemble import RandomForestClassifier

#n_estimators = number of individual decision trees. Higher numbers more accurate but slower
#min_samples_split = helps prevent against overfit. Higher number less accurate but less overfit
#random_state = makes random number tree initialization predictable for model reruns
# - helps with model updates. see if performance change is caused by parameters, not randomness
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

# Initial unsophisticated data split. 100 test rows, rest for training
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

# Ensure we dont accidentally use the tomorrow or target columns. Prevent leakage
predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

Dont use cross validation with time series data because you will be using future data to predict the past which will result in leakage - leaking information into the model (the model will have access to information it wont in real world applications.

```python
# Measure model precision (tp/(tp + fp))
from sklearn.metrics import precision_score
# Metric: When we said it would go up, did it go up?
# Modify metric based on model objective
preds = model.predict(test[predictors])
```

Combine the previous steps into a single function:

```python
def predict(train, test, predictors, model):
    # Fit the Model
    model.fit(train[predictors], train["Target"])
    # Generate Predictions
    preds = model.predict(test[predictors])
    # Combine the model into a series, with name "Predictions"
    preds = pd.Series(preds, index=test.index, name="Predictions")
    # Combine into one data frame
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```

# BackTesting

```python
# start = 2500 - Every trading year has about 250 days -> 10 years of data to train first model
# step = 250 - Train each model for about a year then move to the next year
# Use first 10 years of data to predict 11th year, first 11 to predict 12, ...
# Get predictions for many years, give us confidence in our model
def backtest(data, model, predictors, start=2500, step=250):
    # List of data frames, each frame is the predictions for a single year
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy() # All previous indices
        test = data.iloc[i:(i+step)].copy() # Current index + 250
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions) #Combine all into a single dataframe
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

The S&P went up more often than our model predicted. 53.4% > 52.8%

Therefore, we would have been better off buying every day and selling at night than any suggestion by the model

However, a benefit of backtesting is we are assured that we can more accurately test our model without it being overfit and work to modify parameters to increase its precision

# Increasing Precision

## Adding Predictors

```python
#Using a set of rolling averages. Time horizons of rolling averages:
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    # Rolling average against the horizon
    rolling_averages = sp500.rolling(horizon).mean()
    
    # Create Column based on Horizon
    ratio_column = f"Close_Ratio_{horizon}"
    # Ratio between today's close and average over the last x days.
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    # Sum of the target values over the last x days. Number of days price went up
    # Doesn't include current day - prevents leakage
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500
```

Use the .dropna() function to eliminate data with NaN values. The NaN values originate from a lack of preceding data to the data set making the function unable to complete a rolling average over that time frame. i.e. we cannot calculate a 60 day rolling average for the first data point since we don't have any data points for it to be computed with

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators=100, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    # Fit the Model
    model.fit(train[predictors], train["Target"])
    # Generate probability that a row will be a 0 or 1. Stored in Column 2
    preds = model.predict_proba(test[predictors])[:,1]
    # Set a custom threshold (0.6). 
    # Threshold > 0.5 will decreases days predicted but increases chance price actually increases
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    # Combine the model into a series, with name "Predictions"
    preds = pd.Series(preds, index=test.index, name="Predictions")
    # Combine into one data frame
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```

```python
new_predictors
```

```python
# No longer use Open, Close, High, Low, Volume columns as their values don't indicate 
# stock price movement. Instead its reflected in our new model features / predictors
predictions = backtest(sp500, model, new_predictors) 
```

```python
predictions["Predictions"].value_counts()
```

The model makes significantly fewer predictions that the price will increase.

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

The predictions of the model became more accurate as a result, now being greater than baseline.

57.4% > 54.5%

Therefore, this model has some predictive value.

Given that the model is given only historical prices and time series data this is a compelling result.
